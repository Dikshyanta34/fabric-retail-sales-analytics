```python
sales_silver = spark.table("Sales_Silver")
customers = spark.table("Customers")
products = spark.table("Products")
stores = spark.table("Stores")
date_dim = spark.table("Date_Dimension")
```

```python
sales_gold = sales_silver \
    .join(customers, "CustomerID", "left") \
    .join(products, "ProductID", "left") \
    .join(stores, "StoreID", "left") \
    .join(date_dim, sales_silver.OrderDate == date_dim.Date, "left")
```

```python
display(sales_gold)
```

```python
from pyspark.sql.functions import col

sales_gold = sales_silver \
    .join(customers, "CustomerID", "left") \
    .join(products, "ProductID", "left") \
    .join(stores, "StoreID", "left") \
    .join(date_dim, sales_silver.OrderDate == date_dim.Date, "left") \
    .select(
        sales_silver["*"],
        col("customers.CustomerName"),
        col("customers.Segment"),
        col("customers.City").alias("CustomerCity"),
        col("customers.State").alias("CustomerState"),
        col("products.ProductName"),
        col("products.Category"),
        col("products.SubCategory"),
        col("stores.StoreName"),
        col("stores.City").alias("StoreCity"),
        col("stores.Region"),
        col("Year"),
        col("Month"),
        col("MonthName")
    )
```

```python
spark.sql("DROP TABLE IF EXISTS Sales_Gold")
```

```python
sales_gold.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("Sales_Gold")
```
